```python
import seaborn as sns
data['date'] = pd.to_datetime(data['date'], errors='coerce')
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data = data.drop(columns=['date'])

numeric_data = data.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numeric_data.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Housing Dataset')
plt.show()


price_correlations = correlation_matrix['price'].sort_values(ascending=False)
print("\nFeatures with highest correlation to price:")
print(price_correlations[1:])
```

This correlation matrix illustrates the relationships between various features in the housing dataset. Here's an analysis of the key observations:

1. **Strong Positive Correlations**: Dark red colors indicate strong positive correlations between features, while dark blue colors indicate strong negative correlations.

2. **Price and Square Footage (`sqft_living`)**: There is a strong positive correlation (~0.7) between `sqft_living` and `price`, indicating that larger homes tend to have higher prices.

3. **Price and Grade**: The `grade` feature (quality rating) also has a strong positive correlation with `price` (~0.66), suggesting that higher-quality homes tend to be more expensive.

4. **Above Ground Square Footage (`sqft_above`) and Total Square Footage (`sqft_living`)**: These features are very strongly correlated (~0.88), which makes sense as a larger above-ground area generally means a larger total area.

5. **Bathrooms and Square Footage (`sqft_living`)**: There is a moderately strong positive correlation (~0.58) between the number of bathrooms and the square footage, implying that larger houses often have more bathrooms.

6. **Zip Code**: The `zipcode` feature has low correlation with most other features, suggesting that geographical location (as represented by the zip code) does not significantly affect structural attributes like square footage or the number of rooms.

This analysis helps identify key features that impact house prices, allowing us to focus on these influential variables in building a predictive model, while less relevant features (like `zipcode`) may be given lower importance or even excluded.

# <div style="text-align: center; font-family: Arial, sans-serif; font-size: 24px; color: #333; background-color: #d0e7ff; padding: 10px; border-radius: 8px;">Part 4</div>


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


train_data, validation_data = train_test_split(data, test_size=0.25, random_state=42)

train_features = train_data.drop(columns=['price'])
train_target = train_data['price']
validation_features = validation_data.drop(columns=['price'])
validation_target = validation_data['price']

scaler = MinMaxScaler()

train_features_scaled = scaler.fit_transform(train_features)
validation_features_scaled = scaler.transform(validation_features)

train_features_scaled = pd.DataFrame(train_features_scaled, columns=train_features.columns, index=train_features.index)
validation_features_scaled = pd.DataFrame(validation_features_scaled, columns=validation_features.columns, index=validation_features.index)

train_data_scaled = train_features_scaled.copy()
train_data_scaled['price'] = train_target

validation_data_scaled = validation_features_scaled.copy()
validation_data_scaled['price'] = validation_target

print("\nScaled Training Data:")
print(train_data_scaled.head())
print("\nScaled Validation Data:")
print(validation_data_scaled.head())
```

# <div style="text-align: center; font-family: Arial, sans-serif; font-size: 24px; color: #333; background-color: #d0e7ff; padding: 10px; border-radius: 8px;">Part 5</div>


```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import mean_squared_error


class MLP_OneHiddenLayer(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(MLP_OneHiddenLayer, self).__init__()
        self.hidden = nn.Linear(input_size, hidden_size)
        self.output = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = torch.relu(self.hidden(x))
        x = self.output(x)
        return x


class MLP_TwoHiddenLayers(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2):
        super(MLP_TwoHiddenLayers, self).__init__()
        self.hidden1 = nn.Linear(input_size, hidden_size1)
        self.hidden2 = nn.Linear(hidden_size1, hidden_size2)
        self.output = nn.Linear(hidden_size2, 1)

    def forward(self, x):
        x = torch.relu(self.hidden1(x))
        x = torch.relu(self.hidden2(x))
        x = self.output(x)
        return x


X_train = torch.tensor(train_data_scaled.drop(columns=['price']).values, dtype=torch.float32)
y_train = torch.tensor(train_data_scaled['price'].values, dtype=torch.float32).view(-1, 1)
X_val = torch.tensor(validation_data_scaled.drop(columns=['price']).values, dtype=torch.float32)
y_val = torch.tensor(validation_data_scaled['price'].values, dtype=torch.float32).view(-1, 1)
```

# <div style="text-align: center; font-family: Arial, sans-serif; font-size: 24px; color: #333; background-color: #d0e7ff; padding: 10px; border-radius: 8px;">Part 6</div>


```python
def train_model(model, X_train, y_train, X_val, y_val, epochs=100, learning_rate=0.001):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    for epoch in range(epochs):
        model.train()
        outputs = model(X_train)
        loss = criterion(outputs, y_train)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if (epoch + 1) % 10 == 0:
            val_loss = evaluate_model(model, X_val, y_val)
            print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}, Validation Loss: {val_loss:.4f}')

def evaluate_model(model, X, y):
    model.eval()
    with torch.no_grad():
        predictions = model(X)
        mse = mean_squared_error(y.numpy(), predictions.numpy())
    return mse

input_size = X_train.shape[1]
hidden_size1 = 64
hidden_size2 = 32

print("Training MLP with one hidden layer:")
model_one_hidden = MLP_OneHiddenLayer(input_size=input_size, hidden_size=hidden_size1)
train_model(model_one_hidden, X_train, y_train, X_val, y_val, epochs=1000, learning_rate=0.03)
val_mse_one_hidden = evaluate_model(model_one_hidden, X_val, y_val)
print(f'Validation MSE for one hidden layer MLP: {val_mse_one_hidden:.4f}\n')


print("Training MLP with two hidden layers:")
model_two_hidden = MLP_TwoHiddenLayers(input_size=input_size, hidden_size1=hidden_size1, hidden_size2=hidden_size2)
train_model(model_two_hidden, X_train, y_train, X_val, y_val, epochs=1000, learning_rate=0.005)
val_mse_two_hidden = evaluate_model(model_two_hidden, X_val, y_val)
print(f'Validation MSE for two hidden layers MLP: {val_mse_two_hidden:.4f}')
```

```python
def train_model_with_tracking(model, X_train, y_train, X_val, y_val, epochs=100, learning_rate=0.001):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    train_losses = []
    val_losses = []
    for epoch in range(epochs):
        model.train()
        outputs = model(X_train)
        train_loss = criterion(outputs, y_train)
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        train_losses.append(train_loss.item())
        val_loss = evaluate_model(model, X_val, y_val)
        val_losses.append(val_loss)
        if (epoch + 1) % 100 == 0:
            print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss.item():.4f}, Validation Loss: {val_loss:.4f}')
    return train_losses, val_losses
```

```python

input_size = X_train.shape[1]
hidden_size1 = 64
hidden_size2 = 32

print("Training MLP with one hidden layer:")
model_one_hidden = MLP_OneHiddenLayer(input_size=input_size, hidden_size=hidden_size1)
train_losses_one_hidden, val_losses_one_hidden = train_model_with_tracking(
    model_one_hidden, X_train, y_train, X_val, y_val, epochs=1000, learning_rate=0.06)


print("\nTraining MLP with two hidden layers:")
model_two_hidden = MLP_TwoHiddenLayers(input_size=input_size, hidden_size1=hidden_size1, hidden_size2=hidden_size2)
train_losses_two_hidden, val_losses_two_hidden = train_model_with_tracking(
    model_two_hidden, X_train, y_train, X_val, y_val, epochs=1000, learning_rate=0.005)


plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(train_losses_one_hidden, label='Train Loss')
plt.plot(val_losses_one_hidden, label='Validation Loss')
plt.title('Loss Curves for MLP with One Hidden Layer')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(train_losses_two_hidden, label='Train Loss')
plt.plot(val_losses_two_hidden, label='Validation Loss')
plt.title('Loss Curves for MLP with Two Hidden Layers')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

# <div style="text-align: center; font-family: Arial, sans-serif; font-size: 24px; color: #333; background-color: #d0e7ff; padding: 10px; border-radius: 8px;">Part 7</div>


```python
import numpy as np
print(f"Final Validation MSE for MLP with One Hidden Layer: {val_mse_one_hidden:.4f}")
print(f"Final Validation MSE for MLP with Two Hidden Layers: {val_mse_two_hidden:.4f}")


if val_mse_one_hidden < val_mse_two_hidden:
    best_model = model_one_hidden
    best_model_name = "MLP with One Hidden Layer"
    best_validation_loss = val_mse_one_hidden
else:
    best_model = model_two_hidden
    best_model_name = "MLP with Two Hidden Layers"
    best_validation_loss = val_mse_two_hidden

print(f"\nBest Model: {best_model_name} with Validation MSE: {best_validation_loss:.4f}")


np.random.seed(42)
random_indices = np.random.choice(len(X_val), size=5, replace=False)
selected_samples = X_val[random_indices]
actual_prices = y_val[random_indices]
best_model.eval()
with torch.no_grad():
    predicted_prices = best_model(selected_samples)

print("\nPredictions on 5 Random Validation Samples:")
print("Index | Actual Price | Predicted Price")
for i, (actual, predicted) in enumerate(zip(actual_prices, predicted_prices)):
    print(f"{random_indices[i]:5d} | {actual.item():12.2f} | {predicted.item():14.2f}")
```
