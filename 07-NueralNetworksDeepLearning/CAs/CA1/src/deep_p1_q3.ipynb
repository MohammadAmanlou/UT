Explanation of MRI (Madaline Rule I)
Madaline Rule I (MRI) is one of the earliest algorithms developed for training Madaline networks. It focuses on finding an optimal set of weights for minimizing misclassifications. Here’s a breakdown of how MRI works:

Key Points of MRI:
Error Minimization:

MRI operates by directly minimizing the number of misclassifications. It evaluates the output layer’s response to inputs and adjusts weights when the network misclassifies a sample.
Weight Adjustment Process:

If the output does not match the expected output for a given input, MRI adjusts the weights to correct the misclassification.
It uses a "steepest descent" approach to update weights, meaning it tries to find the smallest weight change that corrects the classification error for the specific training example.
Steps of MRI:

Step 1: Present an input pattern to the network and compute the output.
Step 2: Check if the output matches the desired output (target label). If the output is correct, no change is made.
Step 3: If there is a misclassification, adjust the weights of the output layer. MRI attempts to adjust the weights of each node to bring the network closer to the correct classification.
Step 4: Repeat the process until all samples are correctly classified, or a certain number of iterations are reached.
Limitations of MRI:

MRI is effective for small networks but may not be efficient for larger, more complex datasets, as it does not adaptively adjust for varying error gradients


Explanation of MRII (Madaline Rule II)
Madaline Rule II (MRII) is an advanced version of MRI that overcomes some of its limitations. MRII is more sophisticated, as it uses a heuristic approach to make the training process faster and more effective, especially for networks with multiple layers.

Key Points of MRII:
Pattern-by-Pattern Adjustment:

MRII adjusts weights for each misclassified input, but it considers a wider range of adjustments than MRI. MRII employs a technique known as the “minimal disturbance” principle.
Heuristic Adjustment:

MRII adjusts weights in a way that minimizes the disturbance to other correctly classified patterns. It selectively changes weights only if it will reduce the overall error without negatively impacting other correctly classified patterns.
Steps of MRII:

Step 1: Present an input and calculate the network’s output.
Step 2: If there is a misclassification, MRII identifies which weights need adjustment.
Step 3: Adjust the weights in a way that minimally disturbs the network’s overall structure, aiming to reduce the error while maintaining correct classifications for other patterns.
Advantages of MRII:

MRII is faster and more efficient than MRI, particularly for networks with many layers or larger datasets.
It often achieves higher accuracy by selectively and minimally adjusting weights, making it more adaptable to complex patterns in data.

```python
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import mean_squared_error, accuracy_score, precision_score, recall_score, f1_score
X = data[[0, 1]].values
y = data[2].values
y[y==0] = -1
```

```python
class MRI:
    def __init__(self, n_neurons, learning_rate=0.01, n_iters=75):
        self.n_neurons = n_neurons
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        
        self.output_layer_weights = np.ones(n_neurons) 
        self.output_layer_bias = (n_neurons - 1) 

    def _activation(self, x):
        return np.where(x >= 0.0, 1, -1)
    
    def fit(self, X, Y):
        self.weights = np.zeros((self.n_neurons, X.shape[1])) + 0.001
        self.bias = np.zeros(self.n_neurons) + 0.001
        self.cost_per_epoch = []
        self.accuracy_per_epoch = []

        for _ in tqdm(range(self.n_iters)):
            cost = 0
            for x, label in zip(X, Y):
                z_in = np.dot(self.weights, x) + self.bias
                z = self._activation(z_in)

                y_in = np.dot(z, self.output_layer_weights) + self.output_layer_bias
                y = self._activation(y_in)

                cost += ((label - y) ** 2) / 2.0

                if label == 1 and y == -1:
                    to_update = np.argmax(z_in)
                    self.weights[to_update] += self.learning_rate * (1 - z_in[to_update]) * x
                    self.bias[to_update] += self.learning_rate * (1 - z_in[to_update])

                elif label == -1 and y == 1:
                    to_update = z_in > 0
                    n = np.sum(to_update)
                    if n > 0:
                        x1 = (-1 - z_in[to_update]).reshape(n, 1)
                        x2 = x.reshape(1, -1)
                        self.weights[to_update] += self.learning_rate * x1 * x2
                        self.bias[to_update] += self.learning_rate * (-1 - z_in[to_update])

            self.cost_per_epoch.append(cost)
            y_pred = self.predict(X)
            accuracy = accuracy_score(Y, y_pred)
            self.accuracy_per_epoch.append(accuracy * 100)
        return self

    def predict(self, X):
        predicted = []
        for x in X:

            z_in = np.dot(self.weights, x) + self.bias
            z = self._activation(z_in)

            y_in = np.dot(z, self.output_layer_weights) + self.output_layer_bias
            y = self._activation(y_in)
            predicted.append(y)
        self.preds = np.array(predicted)
        return np.array(predicted)

    def plot_decision_regions(self, X, y, resolution=0.01):

        plt.figure(figsize=(12, 8))
        for i in range(self.n_neurons):
            if self.weights[i][1] != 0:
                slope = -self.weights[i][0] / self.weights[i][1]
                intercept = -self.bias[i] / self.weights[i][1]
                x_vals = np.array([X[:, 0].min() - 1, X[:, 0].max() + 1])
                y_vals = slope * x_vals + intercept
                plt.plot(x_vals, y_vals, linestyle='--', linewidth=2, label=f'Neuron {i+1}')
            else:
                x_vals = np.full(2, -self.bias[i] / self.weights[i][0])
                y_vals = np.array([X[:, 1].min() - 1, X[:, 1].max() + 1])
                plt.plot(x_vals, y_vals, linestyle='--', linewidth=2, label=f'Neuron {i+1}')

        markers = ("s", "x", "o", "^", "v")
        colors = ("red", "blue", "lightgreen", "gray", "cyan")
        cmap = plt.cm.RdYlBu

        x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx1, xx2 = np.meshgrid(
            np.arange(x1_min, x1_max, resolution),
            np.arange(x2_min, x2_max, resolution)
        )
        grid = np.array([xx1.ravel(), xx2.ravel()]).T
        Z = self.predict(grid)
        Z = Z.reshape(xx1.shape)

        plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
        plt.xlim(xx1.min(), xx1.max())
        plt.ylim(xx2.min(), xx2.max())

        for idx, cl in enumerate(np.unique(y)):
            plt.scatter(
                x=X[y == cl, 0],
                y=X[y == cl, 1],
                alpha=0.9,
                c=colors[idx],
                marker=markers[idx],
                label=cl,
                edgecolor='k'
            )

        plt.xlabel("Feature 1", fontsize=14)
        plt.ylabel("Feature 2", fontsize=14)
        plt.legend(loc="upper left", fontsize=12)
        plt.title("Decision Regions", fontsize=16)
        plt.grid(True)
        plt.show()

    def evaluate(self, X, y):
        y_pred = self.predict(X).flatten()

        mse = mean_squared_error(y, y_pred)
        accuracy = accuracy_score(y, y_pred)
        precision = precision_score(y, y_pred, average='macro', zero_division=0)
        recall = recall_score(y, y_pred, average='macro', zero_division=0)
        f1 = f1_score(y, self.preds, average='macro', zero_division=0)

        print(f"Mean Squared Error: {mse}")
        print(f"Accuracy: {accuracy}")
        print(f"Precision: {precision}")
        print(f"Recall: {recall}")
        print(f"F1 Score: {f1}")

    def plot_accuracy(self):
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, self.n_iters + 1), self.accuracy_per_epoch, marker='o', linestyle='-', color='b')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy (%)')
        plt.title('Training Accuracy over Epochs')
        plt.grid(True)
        plt.show()

n_neurons_list = [3, 4, 8]
for n_neurons in n_neurons_list:
    model = MRI(n_neurons=n_neurons)
    model.fit(X, y)
    model.plot_decision_regions(X, y)
    model.evaluate(X, y)
    model.plot_accuracy()
```

```python
class MRII:
    def __init__(self, input_size, hidden_layers):
        self.hidden_layers = hidden_layers
        self.weights = []

        previous_size = input_size
        for hidden_neurons in hidden_layers:
            self.weights.append(np.random.randn(previous_size, hidden_neurons) * 0.1)
            previous_size = hidden_neurons
        self.weights.append(np.random.randn(previous_size, 1) * 0.1)

    def relu(self, x):
        return np.maximum(0, x)

    def relu_derivative(self, x):
        return np.where(x > 0, 1, 0)

    def forward(self, X):
        activations = [X]
        for i in range(len(self.hidden_layers)):
            Z = self.relu(np.dot(activations[-1], self.weights[i]))
            activations.append(Z)
        y_pred = np.dot(activations[-1], self.weights[-1])
        return y_pred, activations

    def train(self, X, y, iters=4000, lr=0.001, batch_size=16):
        accuracies = []
        for iteration in range(iters):
            idx = np.random.randint(0, X.shape[0], batch_size)
            X_batch = X[idx]
            y_batch = y[idx].reshape(-1, 1)

            y_pred, activations = self.forward(X_batch)

            error = y_batch - y_pred
            d_output = error
            self.weights[-1] += lr * np.dot(activations[-1].T, d_output)
            d = d_output
            for i in range(len(self.hidden_layers) - 1, -1, -1):
                d_hidden = np.dot(d, self.weights[i + 1].T) * self.relu_derivative(activations[i + 1])
                self.weights[i] += lr * np.dot(activations[i].T, d_hidden)
                d = d_hidden
            y_pred_train, _ = self.forward(X)
            y_pred_class = np.where(y_pred_train >= 0.5, 1, 0).reshape(-1)
            accuracy = accuracy_score(y, y_pred_class)
            accuracies.append(accuracy * 100)

        return accuracies, iters

    def plot_decision_boundary(self, X, y, title="Decision Boundary"):
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                             np.arange(y_min, y_max, 0.1))
        Z, _ = self.forward(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, levels=[Z.min(), 0.5, Z.max()], alpha=0.3, colors=['purple', 'yellow'])
        plt.contour(xx, yy, Z, levels=[0.5], colors='black', linewidths=2)
        plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
        plt.title(title)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.show()

    def print_metrics(self, X, y):
        y_pred, _ = self.forward(X)
        y_pred_class = np.where(y_pred >= 0.5, 1, 0).reshape(-1)
        y_true = y.reshape(-1)

        mse = mean_squared_error(y_true, y_pred)
        accuracy = accuracy_score(y_true, y_pred_class)
        precision = precision_score(y_true, y_pred_class)
        recall = recall_score(y_true, y_pred_class)
        f1 = f1_score(y_true, y_pred_class)

        print(f"Mean Squared Error: {mse}")
        print(f"Accuracy: {accuracy}")
        print(f"Precision: {precision}")
        print(f"Recall: {recall}")
        print(f"F1 Score: {f1}")

input_size = 2
for hidden_layers in [[3] , [4] , [8]]:
    model = MRII(input_size, hidden_layers)
    model.train(X, y)
    model.plot_decision_boundary(X, y)
    model.print_metrics(X, y)
```
